```python
# Importing th required libraries
import pandas as pd
import qalsadi.lemmatizer
import numpy as np
```

```python
# Choosing 3 columns: ID, Translated Sentences, and the Sentiment columns
# Transfering the datasets from CSV of the testset and the trainset into dataframes

col_list = ["conv_id", "translated_utterance", "context"]
df_train= pd.read_csv("train_arabic_new.txt",encoding='utf-16-le',sep='\t',usecols=col_list)
```

```python
# We then concatenate both dataframes

df_test_arabic = df_train
```

```python
df_test_arabic
```

```python
# We Drop the Nans

df_test_arabic.dropna(inplace=True)
df = df_test_arabic
```

```python
# Define the positive and negative sentiments from the available sentiments.
def change_context(cont):
    positives = ['sentimental','proud','faithful','joyful','grateful','prepared','excited','surprised','confident','nostalgic','caring','trusting','hopeful','content','impressed','anticipating']
    if(cont in positives):
        return 'pos'
    else:
        return 'neg'
```

```python
# Apply the pre-defined method to change sentiments into positive/negative only to increase the accuracy.
df['context']=df['context'].apply(change_context)
df
```

```python
# Lemmatize the dataset sentences using "qalsadi" lemmatizer.
lemmer = qalsadi.lemmatizer.Lemmatizer()
new_col = [lemmer.lemmatize_text(str(x),return_pos=True) for x in df['translated_utterance']]
df['tokenized_utterance'] = new_col
df
```

```python
# Reset the index
df.reset_index(inplace=True,drop=True)
```

```python
# Remove the stopwards after lemmatization
df["tokenized_utterance_filtered"] = np.nan
for i in df.index:
    output = np.array([])
    for j in range(len(df['tokenized_utterance'][i])):
        if isinstance(df['tokenized_utterance'][i][j], tuple):
            if df['tokenized_utterance'][i][j][1] != 'stopword':
                output = np.append(output, df['tokenized_utterance'][i][j][0])
    df["tokenized_utterance_filtered"][i] = output
df["tokenized_utterance_filtered"]
```

```python
# Adding a copy of the available sentences as another column but shifted up once
# This column will be used as a reply 

df["tokenized_utterance_filtered_Responses"] = df['translated_utterance'].shift(-1)
```

```python
df = df.dropna()
df.reset_index(inplace=True,drop=True)
df.head(10)
```

```python
# Removing the last sentence of every conversation from the new column
# As now, the last sentence in the converstation belongs to a different conversation
for i in range(0,len(df)-1):
    if df['conv_id'][i] != df['conv_id'][i+1]:
        df['conv_id'][i] = np.nan

df
```

```python
# Drop Nans, to remove sentences that don't have reply(last sentences in a conversation).

df = df.dropna()
df.reset_index(inplace=True,drop=True)
df.head(10)
```

```python
# Convert the file into CSV so that we can use it without the need to create every time we run.
df.to_csv('tokenized_utterance_train_arabic(trimmed).csv')
```

```python
# Splitting the dataset into train and test datasets
from sklearn.model_selection import train_test_split
reviews = df['tokenized_utterance_filtered'].values
labels = df['context'].values
reviews_train, reviews_test, y_train, y_test = train_test_split(reviews, labels, test_size=0.2, random_state=1000)
```

```python
# Create a count vectorizer of the train dataset
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(tokenizer = lambda x : x, preprocessor=lambda x : x ,ngram_range=(1,1))
X = vectorizer.fit_transform(reviews_train)
print(vectorizer.get_feature_names())
```

```python

X_train = vectorizer.transform(reviews_train)
X_test = vectorizer.transform(reviews_test)
```

```python
# Create a logistic regression model and training it on the training dataset
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(max_iter=10000)
classifier.fit(X_train, y_train)
```

```python
# Test the trained logisitc regression model on the test dataset
accuracy = classifier.score(X_test, y_test)
print("Accuracy:", accuracy)
```

```python
# Running the Chatbot
# First we lemmatize the query
# Second we remove the stopwards from the query
# Third We use the trained model to determine the sentiment of the query
# Fourth We create Tf-idf for the query
# Fifth we create Tf-idf for all the sentences that have the same sentiment as the query
# Sixth We find the cosine similarity between the query and every sentence
# Last We reply by the next available sentence (next row), while making sure that this reply is included in the same conversation.

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
query = ""
while(query!="bye"):
    query = input("can I help you?")
    tokenized_query = lemmer.lemmatize_text(query,return_pos=True)
    tokenized_query_stop_words = np.array([])
    for j in range(len(tokenized_query)):
        if isinstance(tokenized_query[j], tuple):
            if tokenized_query[j][1] != 'stopword':
                tokenized_query_stop_words = np.append(tokenized_query_stop_words, tokenized_query[j][0])
    tokenized_query_stop_words = [tokenized_query_stop_words]
    testVec = vectorizer.transform(tokenized_query_stop_words)
    query_sentiment= classifier.predict(testVec)
    print("The query sentiment is: ", query_sentiment)
    df_same_query_sent = df.loc[df['context'] == query_sentiment[0]]
    vectorizer_tf_idf = TfidfVectorizer(tokenizer = lambda x : x, preprocessor=lambda x : x)
    corpus_tf_idf = vectorizer_tf_idf.fit_transform(df_same_query_sent['tokenized_utterance_filtered'])
    query_tf_idf = vectorizer_tf_idf.transform(tokenized_query_stop_words)
    cosine_similarities = cosine_similarity(corpus_tf_idf,query_tf_idf)
    response_index = np.argmax(cosine_similarities)
    df_same_query_sent.reset_index(inplace=True)
    print("Query is: ", query)
    print("Chatbot: ", df_same_query_sent['tokenized_utterance_filtered_Responses'][response_index])
```
